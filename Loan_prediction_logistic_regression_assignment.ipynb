```python
#Script to find the outliers
for col_name in train.select_dtypes(include=np.number).columns[:-1]:
    print(col_name)
    q1 = train[col_name].quantile(0.25)
    q3 = train[col_name].quantile(0.75)
    iqr = q3 - q1
    
    low = q1-1.5*iqr
    high = q3+1.5*iqr
    print("Outliers = ",train.loc[(train[col_name] < low) | (train[col_name] > high), col_name])
```
